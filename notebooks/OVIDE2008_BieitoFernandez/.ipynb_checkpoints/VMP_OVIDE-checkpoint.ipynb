```python
import xarray as xr
from scipy import io
import numpy as np
import pandas as pd
import seawater as sw
import matplotlib.pyplot as plt
import cmocean
import gsw
#this is just a useful function        
def moving_average(x,n, window = "flat"):
    if n%2 == 0:
        n+=1
    N = x.size
    cx = np.full(x.size, np.nan)
    for i in range(N):
        ii = np.arange(i-n//2, i+n//2+1,1)
        if window == "flat":
            ww = np.ones(ii.size)
        elif window == "gauss":
            xx = ii - i
            
            ww = np.exp(- xx**2/(float(n)/4)**2 )
        elif window == "hanning":
            ww = np.hanning(ii.size)
        ww = ww[ (ii>=0) & (ii<N)]
        ii = ii[ (ii>=0) & (ii<N)]
        
        #print(ii)
        kk = np.isfinite(x[ii])
        if np.sum(kk)<0.25*ii.size:
            continue
        cx[i] = np.sum(x[ii[kk]]*ww[kk])/np.sum(ww[kk])
    return cx
```

# First, load datafiles

```python
dsvmp = io.loadmat("OVIDE2008_processed_VMP6000_final.mat")
dsctd = io.loadmat("OVIDE_VMP_CTD_merged.mat")
print(dsvmp.keys())
#for all variables it only gets data from depth 3 to avoid noise close to the surface
lon = dsvmp["longitude"][0,16:]
lat = dsvmp["latitude"][0,16:]

pres = dsvmp["pres"][0,3:].astype(float)
epsilon = dsvmp["epsilon"][3:,16:]
chi = dsvmp["chi"][3:,16:]

theta = dsvmp["theta"][3:,16:]
salinity = dsvmp["S"][3:,16:]
T = dsvmp["T"][3:,16:]
gamman = dsvmp["gamman"][3:,16:]#dsvmp["gamman"][3:,:]

depth = np.full(theta.shape, np.nan)
for i in range(lon.size):
    for j in range(pres.size):
        depth[j,i] = sw.dpth(pres[j], lat[i])
```

```python
#replaces first ctd profile which was broken (first VMP station was in station 8 -  so index 7)
#T[:,0] = dsctd["temperature"][:,7]
#salinity[:,0] = dsctd["salinity"][:,7]
#gamman[:,0] = dsctd["gamman"][:,7]
#theta[:,0] = sw.ptmp(salinity[:,0], T[:,0], pres,0)
Z = sw.dpth(pres, np.nanmean(lat))

#calculates the vertical gradient
CT = np.full(T.shape, np.nan)
SA = np.full(CT.shape, np.nan)
dCTdz = np.full(CT.shape, np.nan)
dSAdz = np.full(CT.shape, np.nan)
for i in range(lon.size):
    SA[:,i] = gsw.SA_from_SP(salinity[:,i],pres,lon[i],lat[i])
    CT[:,i] = gsw.CT_from_t(salinity[:,i],T[:,i],pres)
    for j in range(1,pres.size-1):
        dCTdz[j,i] = -(CT[j+1,i] - CT[j-1,i])/(depth[j+1,i] - depth[j-1,i])
        dSAdz[j,i] = -(SA[j+1,i] - SA[j-1,i])/(depth[j+1,i] - depth[j-1,i])
```

```python
N2 = np.full(gamman.shape, np.nan)
for i in range(lon.size):
    print(i)
    gamman0 = np.copy(gamman[:,i])
    ii = np.isfinite(gamman0)
    depth00 = depth[ii,i]
    gamman00 = gamman0[ii]
    gamman00 = np.sort(gamman00)
    N20 = np.full(gamman00.size, np.nan)
    N20[1:-1] = 9.81/(1000+gamman00[1:-1])*(gamman00[2:]-gamman00[:-2])/(depth00[2:]-depth00[:-2])
    print(N20)
    N20[0] = N20[1]
    N20[-1] = N20[-2]
    N2[ii,i] = N20
    
```

```python

ds = xr.Dataset(
        data_vars = {
            #grid
            "Longitude" : ( ("station"), lon, {"Description": "Longitude", "Units": "deg"} ),
            "Latitude" : ( ("station"), lat, {"Description": "Latitude", "Units": "deg"} ),
            "depth" : ( ("pressure","station"), depth, {"Description": "Pressure", "Units": "dbar"} ),
            "T" : ( ("pressure","station"), T, {"Description": "Temperature", "Units": "degC"} ),
            "SP" : ( ("pressure","station"), salinity, {"Description": "Salinity practical", "Units": "_"} ),
            "CT" : ( ("pressure","station"), CT, {"Description": "Conservative Temperature", "Units": "degC"} ),
            "SA" : ( ("pressure","station"), SA, {"Description": "Salinity Absolute", "Units": "g/kg"} ),
            "gamman" : ( ("pressure","station"), gamman, {"Description": "Neutral density", "Units": "kg/m3"} ),
            "dCTdz" : ( ("pressure","station"), dCTdz, {"Description": "Conservative Temperature vertical gradient", "Units": "degC/m"} ),
            "dSAdz" : ( ("pressure","station"), dSAdz, {"Description": "Absolute salinity vertical gradient", "Units": "(g/kg))/m"} ),
            "N2" : ( ("pressure","station"), N2, {"Description": "Buoyancy frequency squared", "Units": "1/s^2"} ),
            "epsilon" : ( ("pressure","station"), epsilon, {"Description": "TKE dissipation", "Units": "W/kg"} ),
            "chi" : ( ("pressure","station"), chi, {"Description": "Thermal variance dissipation", "Units": "K^2/s"} ),
            
            
        },
        coords = {
            "pressure": pres,
            "station": np.arange(1,lon.size+1),
        },  
        attrs = {"Description":"OVIDE 2008 microstructure (irminger sea)"},
    )
ds.pressure.attrs['Description'] = "Pressure (dbar)"
ds.station.attrs['Description'] = 'Station'
display(ds)

ds.to_netcdf("OVIDE2008_microstructure.nc")
```

```python

GAMMA_mix = 0.2
chi_dia = 2*dCTdz**2*GAMMA_mix*epsilon/N2
chiS = 2*dSAdz**2*GAMMA_mix*epsilon/N2
chiS[chiS>1e-4] = np.nan #removes some outlier
```

```python
plt.semilogx(moving_average(np.nanmean(chi, axis =1),20),-Z)
plt.semilogx(moving_average(np.nanmean(chi_dia, axis =1),20),-Z)
#plt.semilogx(moving_average(np.nanmean(chi_dia2, axis =1),20),-Z[1:])
```

```python
plt.semilogx(moving_average(np.nanmean(chiS, axis =1),20),-Z)
plt.semilogx(moving_average(np.nanmean(chi_dia, axis =1),20),-Z)#plt.semilogx(moving_average(np.nanmean(chi_dia2, axis =1),20),-Z[1:])
```

```python
dSG = 0.1
SGi = np.arange(25-dSG/2,28.7+dSG/2,dSG)
SG = 0.5*(SGi[1:]+SGi[:-1])

z_sg = np.full(SG.size, np.nan)
p_sg = np.full(SG.size, np.nan)
h_sg = np.full(SG.size, np.nan)

CT_sg = np.full(SG.size, np.nan)
SA_sg = np.full(SG.size, np.nan)

#z_sg = np.full(SG.size, np.nan)
chi_sg = np.full(SG.size, np.nan)
s_chi_sg = np.full(SG.size, np.nan)

chiS_sg = np.full(SG.size, np.nan)
s_chiS_sg = np.full(SG.size, np.nan)

epsilon_sg = np.full(SG.size, np.nan)
s_epsilon_sg = np.full(SG.size, np.nan)

dSAdz_sg = np.full(SG.size, np.nan)
dCTdz_sg = np.full(SG.size, np.nan)
N2_sg = np.full(SG.size, np.nan)
#chi_iso_sg = np.full(SG.size, np.nan)
#s_chi_iso_sg = np.full(SG.size, np.nan)

zz = np.tile(Z,(epsilon.shape[1],1)).T

for i in range(SG.size):
    ii = (np.abs(gamman-SG[i])<dSG/2) & (np.isfinite(chi+chi_dia))
    
    nd = np.sum(ii).astype(float)

    z_sg[i] = np.nanmean(zz[ii])
    p_sg[i] = sw.pres(z_sg[i],np.nanmean(lat))
    CT_sg[i] = np.mean(CT[ii])
    SA_sg[i] = np.mean(SA[ii])

    chi_sg[i] = np.mean(chi[ii])
    s_chi_sg[i] = np.std(chi[ii])/np.sqrt(np.sum(ii).astype(float))
    
    chiS_sg[i] = np.mean(chiS[ii])
    s_chiS_sg[i] = np.std(chiS[ii])/np.sqrt(np.sum(ii).astype(float))

    epsilon_sg[i] = np.mean(epsilon[ii])
    s_epsilon_sg[i] = np.std(epsilon[ii])/np.sqrt(np.sum(ii).astype(float))
    
    iitop =  (np.abs(gamman-(SG[i]-dSG/2))<dSG/2)
    iibottom =  (np.abs(gamman-(SG[i]+dSG/2))<dSG/2)
    
    h_sg[i] = -(np.nanmean(zz[iitop])-np.nanmean(zz[iibottom]))
    dCTdz_sg[i] = (np.mean(CT[iitop])-np.nanmean(CT[iibottom]))/h_sg[i]
    dSAdz_sg[i] = (np.mean(SA[iitop])-np.nanmean(SA[iibottom]))/h_sg[i]
    
    N2_sg[i]=-gsw.grav(np.nanmean(lat),p_sg[i])*( - dCTdz_sg[i]*gsw.alpha(SA_sg[i], CT_sg[i], p_sg[i]) \
                                    + dSAdz_sg[i]*gsw.beta(SA_sg[i], CT_sg[i], p_sg[i]) )
    
    

chi_dia_sg = 2*dCTdz_sg**2*GAMMA_mix*epsilon_sg/N2_sg
s_chi_dia_sg = 2*dCTdz_sg**2*GAMMA_mix*s_epsilon_sg/N2_sg

chiS_dia_sg = 2*dSAdz_sg**2*GAMMA_mix*epsilon_sg/N2_sg
s_chiS_dia_sg = 2*dSAdz_sg**2*GAMMA_mix*s_epsilon_sg/N2_sg
```
